```python
import os
import numpy as np
import re
import datetime
from dateutil.parser import parse
from fy4aMain import FY4AMain
import schedule
import time

pathToFile = "/mnt/FY4A/2019/"
dataTime = "61459"
outputFilePath = "/mnt/FY4A/temp/"

def get_masked_file(pathToFile):
    pattern = ["^(?P<ftype>[A-Z]{2}[0-9]{1}[A-Z]{1})", 
           "[-]{1}[_]{1}",                
           "([A-Z]{4})",                  
           "[-]{2}[_]{1}",                
           "([A-Z]{1})",                  
           "[_]{1}",                      
           "([A-Z]{4})",                  
           "[_]{1}", 
           "([0-9]{4}[A-Z]{1})", 
           "[_]{1}", 
           "([A-Z]{1}[0-9]{1})", 
           "[-]{1}[_]{1}", 
           "([A-Z]{3})", 
           "[-]{1}[_]{1}", 
           "([A-Z]{4})", 
           "[_]{1}", 
           "([A-Z]{3})", 
           "[_]{1}", 
           "(?P<start_data>[0-9]{14})", 
           "[_]{1}", 
           "(?P<stop_data>[0-9]{14})", 
           "[_]{1}", 
           "([0-9]{4}[A-Z]{1})", 
           "[_]{1}", 
           "([A-Z]{1}[0-9]{4})", 
           "[.]{1}", 
           "([A-Z]{3})$"]

    pattern = ''.join(pattern)
    return re.search(pattern, pathToFile)
    

def get_last_data():
    folders = os.listdir(pathToFile)
    files = [os.path.join(pathToFile, file) for file in folders]
    last = max(files, key=os.path.getctime)
    files = os.listdir(last)
    
    fileName = [n for n in files if get_masked_file(n).group("stop_data").endswith(dataTime)]
    lastData = get_masked_file(fileName[0]).group("stop_data")
    
    return str(last)+"/"+str(fileName[0]), lastData

def saveTempImage():
    try:
        lastPath ,lastData = get_last_data()
    
        data = parse(lastData)
        directory = outputFilePath + data.strftime("%Y%m%d")
        if not os.path.exists(directory):
            try:
                os.mkdir(directory)
                print ("Успешно создана директория %s " % directory)
            except OSError:
                print ("Создать директорию %s не удалось" % directory)
        else:
            print ("Директория уже есть %s " % directory)
        
        filename = os.path.basename(lastPath)
        index = filename.index('.')
        ouputFile = directory +"/"+filename[:index]+".tif"
    
        if os.access(ouputFile, os.F_OK) == True:
            print("Файл уже существует: %s" %ouputFile)
            return
        myFY4AMain = FY4AMain()
        myFY4AMain.initParams(
            hdf5_files=lastPath,
            export_files=ouputFile,
            geo_range = "35,65,60,135,0.1",
            resolution = "4000M"
        )
        print("Файл создан: %s " % ouputFile) 
    except Exception as e:
        print(str(e))
        
if __name__ == "__main__":
    saveTempImage()
    print(schedule.every(360).minutes.do(saveTempImage))
    while True:
        schedule.run_pending()
        time.sleep(10)
```
